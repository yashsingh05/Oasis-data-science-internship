
# Import Required Libraries:

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

# Load the Iris Dataset:

```python
iris_data = sns.load_dataset('iris')
```

# Print Dataset Shape:

```python
print(iris_data.shape)
```

# Print Dataset Information:

```python
print(iris_data.info())
```

# Print Summary Statistics:

```python
print(iris_data.describe())
```

# Print Correlation Matrix:

```python
print(iris_data.corr())
```

# Grouping and Printing Species Counts:

```python
print(iris_data.groupby('species').size())
```

# Pairplot with KDE (Kernel Density Estimate):

```python
g = sns.pairplot(iris_data, hue='species', kind='scatter', markers=["o", "s", "D"])
g.map_lower(sns.kdeplot, levels=4, color=".2")
plt.show()
```

# Heatmap of Correlation Matrix:

```python
plt.figure(figsize=(10, 8))
sns.heatmap(iris_data.corr(), cmap="YlOrRd", annot=True)
plt.title('Correlation Matrix', fontsize=20)
plt.show()
```

# Box Plots for Sepal Length vs Species:

```python
plt.figure(figsize=(15, 8))
sns.boxplot(x='species', y='sepal_length', data=iris_data)
plt.title('Sepal Length vs Species', fontsize=20)
plt.show()
```

# Box Plots for Petal Length vs Species:

```python
plt.figure(figsize=(15, 8))
sns.boxplot(x='species', y='petal_length', data=iris_data)
plt.title('Petal Length vs Species', fontsize=20)
plt.show()
```

# Scatter Plot for Sepal Length vs Sepal Width:

```python
plt.figure(figsize=(15, 8))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris_data)
plt.title('Sepal Length vs Sepal Width', fontsize=20)
plt.show()
```

# Scatter Plot for Petal Length vs Petal Width:

```python
plt.figure(figsize=(15, 8))
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=iris_data)
plt.title('Petal Length vs Petal Width', fontsize=20)
plt.show()
```
